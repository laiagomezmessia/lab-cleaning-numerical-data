```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt       
import numpy as np
```

```python
numerical_df = pd.read_csv('numerical.csv')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def distribution_and_transform_plots(dataframe):
    for column in dataframe.columns:
        # Original distribution
        sns.histplot(data=dataframe, x=column, kde=True)
        plt.title(f'Distribution plot: {column}')
        plt.show()
        
        # Log-transformed distribution
        log_transformed_data = np.log1p(dataframe[column])
        sns.histplot(data=log_transformed_data, kde=True)
        plt.title(f'Log-Transformed Distribution plot: {column}')
        plt.show()
        
        # Square root transformation
        sqrt_transformed_data = np.sqrt(dataframe[column])
        sns.histplot(data=sqrt_transformed_data, kde=True)
        plt.title(f'Square Root-Transformed Distribution plot: {column}')
        plt.show()
        %matplotlib inline


# Call the function to create distribution and transformation plots
distribution_and_transform_plots(numerical_df)
```
